```python
import numpy as np
textfile = '/Users/Patti/classes/Python/notebooks/data/asteroids5000.csv'
data= np.genfromtxt(textfile, delimiter=',')
```

```python
n_missing_vals = np.sum(np.isnan(data))
print '# missing values = ',n_missing_vals

n_rows = len(data)
n_incomplete_rows = np.sum(np.isnan(np.sum(data,axis=1)))
n_complete_rows = n_rows - n_incomplete_rows
print '# compelte rows = ',n_complete_rows
```

```python
w = np.where(np.isnan(np.sum(data,axis=1))==0)[0]
new_data = data[w,:]
```

```python
header = open(textfile).readline()[1:-1]
columns = header.split(',')

for i,c in enumerate(columns):
    cmin,cmax,cmean,cstd = np.min(new_data[:,i]),np.max(new_data[:,i]),np.mean(new_data[:,i]),np.std(new_data[:,i])
    print c
    print 'min = ',cmin,', max = ',cmax,', mean = ',cmean,', std = ',cstd
    print'\n'
```
