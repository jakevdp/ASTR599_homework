generate a FITS image of a de Vaucouleurs galaxy (Jaffe profile)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits

# Jaffe Profile
def jaffe(r,r0=1.,rho0=1.):
    return rho0/4/np.pi *(r/r0)**-2*(1.+r/r0)**-2
    

X,Y = np.meshgrid(np.linspace(-100,100,1000),np.linspace(-100,100,1000))
r=np.sqrt(X**2+Y**2)

galaxy = jaffe(r)


plt.imshow(np.log10(galaxy),origin='lower',cmap='binary')
plt.xticks([])
plt.yticks([])
plt.show()

hdu = fits.PrimaryHDU(galaxy)
hdu.writeto('galaxy.fits')



```
