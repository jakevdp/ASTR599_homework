I use the template given in lecture to make a mesh grid of points, then parametrize cartesian coordinates (x, y) with radius r. The de Vaucouleurs profile, or "Jaffe profile" is given as $\rho = \rho_0 / 4\pi^2 (r/r_0)^{-2} (1+r/r_0)^{-2}$, where $r_0$ and $\rho_0$ are some constants. I then write an image deVau.fits. To ensure that the numerical profile follows a de Vaucouleurs surface brightness profile, I plot the fits file in log scale to get a visible contrast.  

```python
%matplotlib inline
import astropy.io.fits as fits
import matplotlib.pyplot as plt
import numpy as np

# make a mesh of points
x, y = np.ogrid[-50:50, -50:50]
x.shape, y.shape
r = np.sqrt(x**2 + y**2)
rho0 = 1.
print r.shape
print np.min(r), np.max(r)
# Use the 'Jaffe profile' to simulate de Vaucouleurs galaxy luminosity profile
# r0 is 1/e of the radial extent of the galaxy
r0 = np.max(r)*.37
# Jaffe profile
rho = rho0 / (4.*np.pi**2) * (r/r0)**(-2) * (1. + r/r0)**(-2)
#rho = rho0 / (4.*np.pi**2) * (r0 / (r * (1. + r0))**2)
hdu = fits.PrimaryHDU(rho)
hdu.writeto('deVau.fits')
```

```python
# now plot to see that it's actually a de Vaucouleurs profile
# must use log scaling for contrast
import matplotlib as M
plt.figure()
plt.imshow(rho, plt.cm.gray, aspect='equal',norm=M.colors.LogNorm())
```
