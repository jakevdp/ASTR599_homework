```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
```

```python
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

Seems like the SVC method is not a good method to apply to this dataset. Try 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
```

```python
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

The Random Forest Classifier method seems to classify RR Lyrae stars with better precision (79% compared with 65% for K-Nearest Neighbors method). The completeness (recall) drops from 60% in the case of K-Nearest Neighbors to 46%, however. 

```python
from sklearn.lda import LDA
clf = LDA()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
```

```python
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

Although more complete, this method gave poor precision when classifying RR Lyrae variables. The most precise classifier method appears to be the Random Forest Classifier.

```python
from astroML.datasets import fetch_sdss_specgals
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
from sklearn.neighbors import KNeighborsRegressor
est = KNeighborsRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

The RMS has improved by (0.0323 - 0.0235) / (0.0323+0.0235) * 200 = 31% when using the K-Nearest Neighbor Regressor instead of Linear Regressor.

```python
from sklearn.svm import SVR
est = SVR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

The SVR method didn't do a great job at fitting photometric redshifts.

```python
from sklearn.ensemble import GradientBoostingRegressor
est = GradientBoostingRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

The Gradient Boosting Regressor method was comparable to the K-Neighbors Regressor, but much slower at learning.
