I picked '5895287' as my variable star. I utilize the Lomg-Scargle periodigram to find the peak period frequency as was done in class. 

```python
%matplotlib inline
import astroML.datasets
import numpy as np
import matplotlib.pyplot as plt
lcs = astroML.datasets.fetch_LINEAR_sample()
time, flux, dflux = lcs[5895287].T
plt.errorbar(time, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("MJD")
plt.ylabel("Mag")
plt.title("Light Curve")
plt.show()
```

```python
from astroML.time_series import lomb_scargle
periods = np.logspace(-1, 0, 100000, base=10)
periodogram = lomb_scargle(time, flux, dflux, omega=2 * np.pi / periods, generalized=True)
plt.plot(periods, periodogram)
plt.semilogx()
plt.xlabel("Period (days)")
plt.ylabel("Power")
plt.show()
```

```python
idx = np.argsort(periodogram)
print "Period, Power: ",periods[idx[-1]], periodogram[idx[-1]]
period = periods[idx[-1]]
phase = time / period - time // period
```

After exploring the period space I found that P = 0.5392 days corresponds to the highest power. I do a visual check to make sure the period is not an alias signal. Indeed, the light curve folded to this period looks well behaved. 

```python
plt.errorbar(phase, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.title("Folded Light Curve")
plt.show()
```

```python
# using the Fourier class written by A. Becker
# note nterms = 2 ==> 3 parameters (flux + pair)
# nterms = 3 ==> 5 parameters (flux + pair + pair)
import scipy.optimize as opt
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms): 
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
    
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess
```

```python
nterms     = np.arange(1, 30, 2)
chi2       = []
BICpenalty = []
AICpenalty = []
for nterm in nterms:
    fourier = Fourier(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    BICpenalty.append(len(guess) * np.log(len(fourier.phase)))
    AICpenalty.append(len(guess) * 2.0)
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
```

Below I plot chi^2, BIC, AIC vs. nu (# of parameters), as well as delta chi^2 vs. nu.

```python
plt.plot(nterms*2-1, chi2, label=r'$\chi^2$')
plt.plot(nterms*2-1, chi2 + BICpenalty, label="BIC")
plt.plot(nterms*2-1, chi2 + AICpenalty, label="AIC")
plt.xlabel(r'$\nu$')# parameters (Nterms*2-1)")
plt.ylabel(r'$\chi^2$')
plt.semilogy()
plt.legend()
plt.show()
```

```python
dchi2=chi2[:-1]-chi2[1:]
plt.plot(nterms[1:]*2-1, dchi2,"bo-")
plt.xlabel(r'$\nu$')
plt.ylabel(r'$\Delta \chi^2$')
plt.semilogy()
plt.show()
```

```python
bic,aic,dch = np.argsort(chi2+BICpenalty)[0],np.argsort(chi2+AICpenalty)[0], np.argsort(dchi2)[0]
print "Model order (min AIC, BIC, ∆chi^2): ", aic, bic, dch
print "Nu parameters (min AIC, BIC, ∆chi^2): ", nterms[aic]*2-1, nterms[bic]*2-1, nterms[dch]*2-1
print nterms[1:]*2-1, dchi2
```

The BIC penalty test suggests the 4th order model (i.e., Nterms = 9; Nu parameters = Nterms*2-1 = 17) is the best. The AIC results suggest that the 12th order model minimizes penalty (i.e., Nterms = 25; nu=49). To determine which order model is best, I calculated the ∆chi^2 values between consecutive models. The plot (∆chi^2 vs. nu parameters) shows the change in minimized chi^2 value for every increase in model complexity (or number of parameters to be fitted). I used a Chi^2 Distribution Table to look up the significance of added model complexity. The difference between a 3rd and 4th order model is significant (Nu = 17 parameters, ∆chi^2 = 50, p-value < 0.001). However, the difference between the 4th and 5th order model is no longer significant (Nu = 25, ∆chi^2 = 4.2, p > 0.995). The higher order models contain a larger number of free parameters, but their ∆chi^2 values are comparable to ∆chi^2 of the 5th order model. Therefore, a 4th order model is sufficient and its improvement upon the previous model is significant. 

Just to be sure, I plot the best-fit models according to BIC (4th order, yellow line) and AIC (12th order, blue line) with the data points below. Indeed, both models follow the general shape of the folded light curve closely. However, since the 4th order model is the most "parsimonious" model, it is chosen for final analysis.
