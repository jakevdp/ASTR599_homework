#### Exercise 1: Classification of RR Lyrae

Goal: Use another classifier and try to improve on this precision/recall score.

```python
from astroML.datasets import fetch_rrlyrae_combined
# Loads the RR Lyrae data set

from sklearn.cross_validation import train_test_split
# Loads the Scikit-Learn function that allows you to 
#     Cut a data set into a training set and a test set

X, y = fetch_rrlyrae_combined()
X_train, X_test, y_train, y_test = train_test_split(X,y)
# values in matrix X contain all color information.
#     Apparently there's 4 columns.  Are these 4 different colors?
#     From the plot in the notes: col 0 = u-g, 1 = g-r.
#     Perhaps 2 = r-i, 3 = i-z?
# values in array y contain what appear to be type information.
#     The types are either 0 or 1.  0 = Main Sequence; 1 = RR Lyrae  
```

```python
# Just testing some stuff about the data set.
# Find out what we've got
print X.shape
print set(y)
```

```python
# Import the basic modules that I always work with
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N_plot = 10000
plt.scatter(X[-N_plot:,0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('SDSS u-g')
plt.ylabel('SDSS g-r')
plt.show()
```

```python
from sklearn.svm import SVC
clf = SVC()
# I suppose we'll just initialize this for now...  
# I'm always fairly worried when I'm black-boxing something though.
```

```python
# Fit on the training set and predict on the test set
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
# Let's see how many objects we classify correctly
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
```

```python
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred, 
                            target_names=['MS star', 'RR Lyrae'])
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

```python
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=2)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

```python
# Wow, KMeans really sucked.  Let's see what happened.
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred, edgecolor='none')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

```python
from sklearn.ensemble import ExtraTreesClassifier
clf = ExtraTreesClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

```python
from sklearn.lda import LDA
clf = LDA()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print classification_report(y_test, y_pred,
                            target_names=['MS star','RR Lyrae'])
```

#### Exercise 2: Photometric Redshifts via Regression

Goal: Try to improve on the example's regression result by using another regression algorithm.

```python
from astroML.datasets import fetch_sdss_specgals
data = fetch_sdss_specgals()
```

```python
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
rms = np.sqrt(np.mean((y_test - y_pred)**2))
print rms
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
from sklearn.ensemble import RandomForestRegressor
est = RandomForestRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
rms = np.sqrt(np.mean((y_test - y_pred)**2))
print rms
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
est = KNeighborsRegressor(n_neighbors=5)
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
rms = np.sqrt(np.mean((y_test - y_pred)**2))
print rms
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.show()
```

```python
from sklearn.svm import SVR
est = SVR()
est.fit(X_train[:], y_train[:])
y_pred = est.predict(X_test)
rms = np.sqrt(np.mean((y_test - y_pred)**2))
print rms
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
est = GradientBoostingRegressor()
est.fit(X_train[::10], y_train[::10])
y_pred = est.predict(X_test)
rms = np.sqrt(np.mean((y_test - y_pred)**2))
print rms
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.show()
```
