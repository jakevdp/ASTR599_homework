#### The homework concerning the Mandelbrot Set.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the
    candidate is in the mandelbrot set for the given
    number of iterations.
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iter

def create_fractal_python(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """
    Create and return a fractal image
    """
    image = np.zeros((Ny,Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
image = create_fractal_python(600, 400, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

```python
%timeit create_fractal_python(600, 400, -2, 1, -1, 1, 20)
```

So, I tried using cython on the "create_fractal" function itself, but it gave me errors with "np.zeros".  So I just did it with "mandel" isntead.  Now to do it with type annotations.

```python
%%cython
def mandel(x, y, max_iter):
    c = complex(x, y)
    cdef complex z = 0.0j
    
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iter
```

```python
%timeit create_fractal_python(600, 400, -2, 1, -1, 1, 20)
```

Woah!  Just declaring "z" as being complex reduced the runtime by about 73%.  Crazy!

```python
%%cython
import numpy as np
cimport numpy as np

cdef mandel(x, y, max_iter):
    cdef complex c = complex(x, y)
    cdef complex z = 0.0j
    
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iter

def create_fractal_cython(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    cdef np.ndarray image = np.zeros((Ny,Nx), dtype=float)
    cdef float dx = (xmax - xmin) * 1. / Nx
    cdef float dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image

```

### Let's Try This With Numba

```python
from numba import autojit
```

```python

def mandel(x, y, max_iter):
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iter
mandel = autojit(mandel)

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    image = np.zeros((Ny,Nx), dtype=np.float64)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
create_fractal_numba = autojit(create_fractal)
```

```python
%timeit create_fractal_numba(600, 400, -2, 1, -1, 1, 20) 
```

### Comparing Run Times

```python
print "Basic Python"
%timeit create_fractal_python(600, 400, -2, 1, -1, 1, 20) 

print "Cython"
%timeit create_fractal_cython(600, 400, -2, 1, -1, 1, 20) 

print "Numba"
%timeit create_fractal_numba(600, 400, -2, 1, -1, 1, 20) 
```
