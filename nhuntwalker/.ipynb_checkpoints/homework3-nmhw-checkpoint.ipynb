Let's make our Fourier class now.

```python
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms):
        """ Constructor module """
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms - 1):
            model += terms[2*i + 1]*np.cos(2*np.pi*(i+1)*phase+terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
```

```python
fourier = Fourier(phase, flux, dflux, 2)
mphase = np.arange(0, 1, 0.01)
model = fourier.evaluate(mphase, [10,0.1,np.pi])
```

Let's add a method to make guesses

```python
class Fourier2(Fourier):
    def __init__(self, phase, flux, dflux, nterms):
        Fourier.__init__(self, phase, flux, dflux, nterms)
        
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2*np.pi*(1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2*np.pi*(1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess
```

```python
nterms = np.arange(1,20)
chi2 = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess = fourier.makeGuess()
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
plt.plot(nterms, chi2)
plt.semilogy()
plt.xlabel("N Terms")
plt.ylabel("Chi$^2$")
plt.show()
```

Let's see the change in $\chi^2$ with the addition of terms

```python
delta_chi2 = chi2[:-1] - chi2[1:]
plt.plot(nterms[1:],delta_chi2)
plt.xlabel("N terms")
plt.ylabel("$\Delta\chi^2$")
plt.show()
```

AIC/BIC penalties

```python
nterms = np.arange(1, 30, 2)
chi2 = []
BICpen = []
AICpen = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess = fourier.makeGuess()
    BICpen.append(len(guess)*np.log(len(fourier.phase)))
    AICpen.append(len(guess)*2.0)
    chi2.append(opt.fmin_bfgs(fourier.chi2,x0=[guess],disp=0,full_output=1)[1])

chi2 = np.array(chi2)
BICpen = np.array(BICpen)
AICpen = np.array(AICpen)
```

```python
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + AICpen, label="AIC")
plt.plot(nterms, chi2 + BICpen, label="BIC")
plt.xlabel("N Terms")
plt.ylabel("$\chi^2$")
plt.semilogy()
plt.legend()
plt.show()
```

Interesting.  It doesn't appear to diverge nearly as obviously as the example from class.  I would personally stop at 5-6 terms, as that seems to be where the $\chi^2$ value appears to plateau.

```python
print np.argsort(chi2+BICpen)[0],np.argsort(chi2+AICpen)[0]
```

This is telling me that I haven't yet minimized my AIC/BIC penalty value.  HOWEVER, I think that I'll stick to stopping at 5-6 terms. And while more terms will give me a sinusoid that fits increasingly better, it'll only be marginally so, and will introduce unnecessary complexity into my model of the object's variability, which isn't reconcilable when trying to find a physical interpretation for an N terms model.
