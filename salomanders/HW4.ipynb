```python
%pylab inline
import astropy.io.fits as fits
import numpy as np
import matplotlib.pyplot as plt

imsize = 50.
x, y = np.ogrid[-imsize:imsize, -imsize:imsize]
x.shape, y.shape
n = np.sqrt(x**2 + y**2)
n.shape # we have a 2D array!
plt.imshow(n)
B = 20.
C = 5.
D = 2.5
galaxy = B*np.exp(-n/20) + C*np.exp(-n/30) + D*np.exp(-n/40)

plt.imshow(galaxy)
plt.plot(galaxy[imsize,:], 'k-')
plt.xlim(0, 100)
plt.ylim(0,100)
hdu = fits.PrimaryHDU(n)
hdu.writeto('ExponentialGalaxy.fits')
```
