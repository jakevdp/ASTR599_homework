```python
import astropy.io.fits as fits
import numpy as np
#test 2D version
#declare image grid
x, y = np.ogrid[-250:250, -250:250]
x.shape, y.shape

#constants for jaffee profile (arbitrary)
p0=100.
r0=30.
jaffe = (r0/(4.*np.pi)) * (np.sqrt(x**2. + y**2.)/r0)**-2. * (1. + np.sqrt(x**2. + y**2.)/r0)**-2.
jaffe.shape #2D array?
if os.path.exists('Jaffe_deVaucouleurs.fits'):
    os.remove('Jaffe_deVaucouleurs.fits')
hdu = fits.PrimaryHDU(jaffe)
hdu.writeto('Jaffe_deVaucouleurs.fits')
#check it out in ds9 - looks okay except center point is a nan
#is this right?
```
