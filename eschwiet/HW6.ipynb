```python
#First we'll reproduce the example exercise from the notebook. Then we'll do the two homework exercises. 
#Exercise from notebook
import numpy as np
import matplotlib.pyplot as plt
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# start the inline backend for plotting
%matplotlib inline
```

```python
#plot main sequence and LL Lyrae stars
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
#try K neighbors classification
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
```

```python
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

```python
#we see above that the K neighbors classification only does a so-so job identifying the RR Lyrae stars. 
#This is an unbalanced set because there are far more MS stars than RR Lyrae stars. 
```

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
#compares photometric redshift (from the regression) to spectroscopic redshift
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
#RMS is lower than you'd think because points are concentrated at low z where fit isn't as bad 
```

```python
#Exercise 2 - Try to improve on this result using another regression algorithm.
#I'm going to try three of them.
```

```python
#The K neighbors regressor is supposed to be relatively inaccurate, so let's try that one first and then compare.
#We'll use the defaults
from sklearn.neighbors import KNeighborsRegressor as KNR
est = KNR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
#compares photometric redshift (from the regression) to spectroscopic redshift
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
#Now that actually looks a lot better than our linear regression!
#Check RMS error
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
#A noticeable improvement!
```

```python
#Let's try RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor as RFR
est = RFR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
#compares photometric redshift (from the regression) to spectroscopic redshift
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
#Can't really tell visually whether this is better than the Kneighbors regression
#Check RMS error
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
#It is marginallly better than KNR!
```
