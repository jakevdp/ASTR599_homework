```python
#PART 1
#1 Use np.genfromtxt to load the data from the file.
import numpy as np
fname='./data/asteroids5000.csv'
data=np.genfromtxt(fname, delimiter=',', comments='#',names=['a','e','i','H','diameter','UV','UB'])
```

```python
#2 - how many values are missing in this data?
#This is probably inefficient but let's count each component of the data separately and add
x1=np.isnan(data['a']).sum()
x2=np.isnan(data['e']).sum()
x3=np.isnan(data['i']).sum()
x4=np.isnan(data['H']).sum()
x5=np.isnan(data['diameter']).sum()
x6=np.isnan(data['UV']).sum()
x7=np.isnan(data['UB']).sum()
no_nans=x1+x2+x3+x4+x5+x6+x7
#pretty sure the above can be done in one line, but I don't know how
print "The number of missing values for a:", x1, "for e:", x2, "for i:",x3,"for H:",x4,"for diameter:",x5,"For UV:",x6,"For UB:",x7
print "The total number of missing values is:",no_nans

```

```python
#2 - How many complete rows are there? i.e. how many objects have no missing values?
#know it has to be less than 5000-4100 = 900 from above
ncomplete=(data == data).sum() #nans don't equal themselves
print "There are", ncomplete, "complete rows of data"
```

```python
#Create a new array containing only the rows with no missing values.
#use a mask here
complete_data=data[(data == data)]
#double check with above
print len(complete_data)
```

```python
#success!
#4 - Compute the maximum, minimum, mean, and standard deviation of the values in each column.
#okay here we're just going to use numpy's built in stuff
#I assume this means with the new, complete array rather than the entire array

#semi-major axis
max_a=np.max(complete_data['a'])
min_a=np.min(complete_data['a'])
mean_a=np.mean(complete_data['a'])
std_a=np.std(complete_data['a'])
print "Semi-major axis - ", "Max:",max_a,"Min:", min_a,"Mean:", mean_a,"STD:", std_a

#eccentricity
max_e=np.max(complete_data['e'])
min_e=np.min(complete_data['e'])
mean_e=np.mean(complete_data['e'])
std_e=np.std(complete_data['e'])
print "Eccentricity - ", "Max:",max_e,"Min:", min_e,"Mean:", mean_e,"STD:", std_e

#inclination
max_i=np.max(complete_data['i'])
min_i=np.min(complete_data['i'])
mean_i=np.mean(complete_data['i'])
std_i=np.std(complete_data['i'])
print "Inclination - ", "Max:",max_i,"Min:", min_i,"Mean:", mean_i,"STD:", std_i

#H - Brightness
max_h=np.max(complete_data['H'])
min_h=np.min(complete_data['H'])
mean_h=np.mean(complete_data['H'])
std_h=np.std(complete_data['H'])
print "H - Brightness (magnitudes) - ", "Max:",max_h,"Min:", min_h,"Mean:", mean_h,"STD:", std_h

#diameter
max_d=np.max(complete_data['diameter'])
min_d=np.min(complete_data['diameter'])
mean_d=np.mean(complete_data['diameter'])
std_d=np.std(complete_data['diameter'])
print "Diameter - ", "Max:",max_d,"Min:", min_d,"Mean:", mean_d,"STD:", std_d

#UV
max_uv=np.max(complete_data['UV'])
min_uv=np.min(complete_data['UV'])
mean_uv=np.mean(complete_data['UV'])
std_uv=np.std(complete_data['UV'])
print "UV - ", "Max:",max_uv,"Min:", min_uv,"Mean:", mean_uv,"STD:", std_uv

#UB
max_ub=np.max(complete_data['UB'])
min_ub=np.min(complete_data['UB'])
mean_ub=np.mean(complete_data['UB'])
std_ub=np.std(complete_data['UB'])
print "UB - ", "Max:",max_ub,"Min:", min_ub,"Mean:", mean_ub,"STD:", std_ub

#I'm SURE there's a better way than this, oh well...
```

```python
#2 - Invoke the pylab inline magic to make figures appear inline in the notebook
import matplotlib.pyplot as plt
%pylab inline
```

```python
#3 - Use plt.scatter to plot the semi-major axis versus the sine of the inclination angle
#make sure to have degrees to radians conversion
plt.scatter(complete_data['a'],np.sin(complete_data['i']*(np.pi/180.)))
plt.xlabel('semi-major axis [AU]')
plt.ylabel('Sine(inclination)')
```

```python
# 3 cont'd What do you notice about the distribution? What do you think could explain this?
#Some narrow ranges in semi-major axes are depopulated in asteroids.
#The explanation is that resonances with Jupiter make these locations unstable for long periods.
```

```python
#4 - Use plt.scatter to plot a color-magnitude diagram of the asteroids (H vs B-V). 
plt.scatter(complete_data['UV'],complete_data['H'])
plt.gca().invert_yaxis()
plt.xlabel('UV')
plt.ylabel('H')
```

```python
#I'm a bit confused about the colors here because of the labeling, is B-V 'UB'-'UV'?
#because that plot does not produce two populations, see
plt.scatter((complete_data['UB']-complete_data['UV']),complete_data['H'])
plt.gca().invert_yaxis()
plt.xlabel('UB-UV')
plt.ylabel('H')
```

```python
#While plotting H vs. UV does produce the two-population scatterplot, so I'll stick to that.
#4 -cont'd You should see two distinct "families" of asteroids indicated in this plot. 
#Over-plot a line that divides these.
plt.scatter(complete_data['UV'],complete_data['H'])
plt.gca().invert_yaxis()
plt.plot([0.78,0.78],[14,2])
plt.xlabel('UV')
plt.ylabel('H')
```

```python
#5 - Repeat the orbital parameter plot from above, but plot the two "families" from #4 in different colors
#make a mask that makes anything with a UV above 0.78 blue and below red
group1=complete_data[(complete_data['UV'] > 0.78)]
group2=complete_data[(complete_data['UV'] < 0.78)]
plt.scatter(group1['a'],np.sin(group1['i']*(np.pi/180.)),color='blue')
plt.scatter(group2['a'],np.sin(group2['i']*(np.pi/180.)),color='red')
plt.xlabel('semi-major axis [AU]')
plt.ylabel('Sine(inclination)')
```

```python
#5 - cont'd - Do you see any correlation between color and orbit?
#Yes, it appears that bluer objects have smaller semi-major axes. 
#Color is a rough proxy for composition, so this may be a result of the condensation sequence
#in the Early Solar System. 
```
