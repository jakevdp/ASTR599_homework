#### Here I'm generating an intensity map of an Elliptical galaxy and saving it as a fits file.

```python
x, y = np.ogrid[-100:100:1, -100:100:1]
Ie = 1.0    #central intensity
xc = -11.1   #offset the center of galaxy, just for fun
yc = 2.67
Re = 1000.0   #half light radius
#generate an elliptical galaxy intensity map, with some noise
Ir = Ie * np.exp(-7.669 *((np.sqrt((x-xc)**2.0+((y-yc-0.7*x)/0.5)**2.0)/Re)**(1./4) -1))+np.random.randn(len(x),len(y[0]))
plt.imshow(np.log10(Ir.T), interpolation = 'nearest', 
            origin ='lower', cmap = 'Greys',extent=[-50,50,-50,50])
plt.colorbar()
hdu = fits.PrimaryHDU(Ir)
hdu.writeto('EllGalaxy.fits')  #save as fits file
```

#### Load it, make sure it looks the same

```python
new_image = fits.open('EllGalaxy.fits')
plt.imshow(np.log10(new_image[0].data), interpolation = 'nearest',
            origin ='lower', cmap = 'Greys',extent=[-50,50,-50,50])
plt.colorbar()
plt.savefig('EllGalaxy.png')
```
