#### Exercise 1: Classification

```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split
X, y = fetch_rrlyrae_combined()
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

###### K nearest neighbors, for comparison

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred, target_names=['MS star','RR Lyrae'])
```

###### Let's try support vectors:

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred, target_names=['MS star','RR Lyrae'])
```

###### Ok, how about Random Forest:

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred, target_names=['MS star','RR Lyrae'])
```

#### Exercise 2: Regression

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

###### Linear regression for comparison

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

###### Let's try K neighbors again:

```python
from sklearn.neighbors import KNeighborsRegressor
est = KNeighborsRegressor(n_neighbors=5)
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',r')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

###### Support Vectors again

```python
from sklearn.svm import SVR
est = SVR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',c')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```
