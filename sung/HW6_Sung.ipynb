```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split
X, y = fetch_rrlyrae_combined()
X_train, X_test, y_train, y_test = train_test_split(X, y)
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
```

```python
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

```python
#Regression
from astroML.datasets import fetch_sdss_specgals
# Assigning Data
data = fetch_sdss_specgals()
# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']
# Split into train_test_splitining and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, '*k')
plt.plot([0, 1], [0, 1], '-k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
#K-Neighbor
from sklearn.neighbors import KNeighborsRegressor as KNR
est = KNR(n_neighbors=5)
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, '*r')
plt.plot([0, 1], [0, 1], '-b')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
#sklearn.svm.SVR (potentially accurate, but slow)
```

```python
from sklearn.svm import SVM
est = SVM()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, '*b')
plt.plot([0, 1], [0, 1], '-r')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
#sklearn.ensemble.RandomForestRegressor (fast, and accurate with tuning)
from sklearn.ensemble import RandomForestRegressor as RFR
est = RFR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, '*c')
plt.plot([0, 1], [0, 1], '-k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
#sklearn.ensemble.GradientBoostingRegressor (accurate with tuning, can be very slow depending on parameters)
from sklearn.ensemble import GradientBoostingRegressor as GBR
est = GBR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, '*y')
plt.plot([0, 1], [0, 1], '-c')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```
