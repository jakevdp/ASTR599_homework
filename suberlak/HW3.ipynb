Now, define a class fitting the Fourier series, and returning the $\chi^2$  for a model fit. The method evaluate() returns the lightcurve for a given set of phases. The model is shown below:

Model = A$_0$ + A$_1$ * cos(2 $\pi$ phase + $\phi_1$) + A$_2$ * cos(2 $\pi$ 2 $\cdot$ phase + $\phi_2$) + ... + A$_N$ * cos(2 $\pi$ N $\cdot$ phase + $\phi_N$)

with model parameters [ A$_0$ A$_1$ $\phi_1$ A$_2$ $\phi_2$ ... A$_N$ $\phi_N$ ]

```python
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms): # define the variables
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):  # define the method : evaluating lightcurve for a chosen number of terms of the Fourier series
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):  # calculate the chi2 statistic for the plot involving a chosen number of terms 
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
```

See how this class works: we call it (create its instance) with just one term: nterms=1.

We see that it picks up noise or periodicity that is not even there.  A way to make the process of adding more terms to the Fourier series more automatic, is to define a Class which inherits from the main definition of Fourier function which we are fitting. 

```python
class Fourier2(Fourier):
    def __init__(self, phase, flux, dflux, nterms):
        Fourier.__init__(self, phase, flux, dflux, nterms)
                
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess
```

Running it for a range of models shows that going from $n=1$ to $n=2$ we  have a biggest improvement in $\chi^2$ value, further adding terms makes only very small improvement. $\chi^2$ will always decrease, i.e. improve the fit, as we add more terms, i.e. provide more degrees of freedom, but this also makes the information much less useful as overparametrised models become degenerate. This means that many different parametrisations of the same overparametrised mode could fit exactly the same data. The way to tell when to stop adding further terms is to quantify how adding further terms improves the goodness of fit.  

```python
nterms = np.arange(1, 10)
chi2   = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    print guess
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
plt.plot(nterms, chi2)
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.show()
    
```

Delta $\chi^2$ provides information about the relative improvement of our fit as we add more terms to the model. Those small jumps around $n=5$ or $n=7$ could mean that some noise, otherwise ignored, suddenly gets interpreted as another peak of a n-term model, and it makes the fit worse since it ignore other data points. 

```python
delta_chi2 = chi2[:-1] - chi2[1:]
plt.plot(nterms[1:], delta_chi2)
plt.xlabel("N terms")
plt.ylabel("Delta chi2")
plt.show()
```

The quantitative way to express the validity of adding more terms versus improving the fit is to calculate Bayesian Information Criterion or Akaike Information Criterion: 

BIC = $\chi^2$ + *Nterm* $\cdot$ ln(*Ndata*)

AIC = $\chi^2$ + *Nterm* $\cdot$ 2 

Thus we calculate those values below:

```python
nterms     = np.arange(1, 10)
chi2       = []
BICpenalty = []
AICpenalty = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    BICpenalty.append(len(guess) * np.log(len(fourier.phase)))
    AICpenalty.append(len(guess) * 2.0)
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
```

```python
print chi2
print BICpenalty
print AICpenalty
```

And plot them : 

```python
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + BICpenalty, label="BIC")
plt.plot(nterms, chi2 + AICpenalty, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()
```

We can see that BIC and AIC are increasing, i.e. telling that there is a penalty involved with trying to overparametrise the model . This can be made more apparent when we try fitting models with even more terms than 9, eg up to $n=20$ : 

```python
nterms     = np.arange(1, 30, 2)
chi2       = []
BICpenalty = []
AICpenalty = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    BICpenalty.append(len(guess) * np.log(len(fourier.phase)))
    AICpenalty.append(len(guess) * 2.0)
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
```

```python
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + BICpenalty, label="BIC")
plt.plot(nterms, chi2 + AICpenalty, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()
```
