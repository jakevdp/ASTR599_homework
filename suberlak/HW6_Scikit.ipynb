### Exercise 1: Classification of RR Lyrae

"Use another classifier and try to improve on this precision/recall score.

There are several you might try:

- ``sklearn.lda.LDA`` (slow but inaccurate)
- ``sklearn.svm.SVC`` (fast but accurate)


First, I import the dataset:

````python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
````

````python
import matplotlib.pyplot as plt
%pylab inline

N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
````

This is the same plot as in the lecture slides. Now, in the similar way to how it was done in the lecture slides, I import another Classifiers, as suggested above.


````python
# 1  
# from http://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html#sklearn.naive_bayes.GaussianNB

from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)

from sklearn.metrics import classification_report
print '\n sklearn.neighbors.GaussianNaiveBayes \n'
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
plt.scatter(X_test[:,0], X_test[:,1], c=y_pred, edgecolor='none')
plt.show()
````

So I get 0.17 and 0.84 as precision and recall respectively for the Naive Bayes Gaussian estimator.

Here I obtain 0.15 and 0.66 for precision and recall, respectively, for the Linear Discriminant Analysis.

### Exercise 2 : Linear Regression

"Try to improve on this result using another regression algorithm.  Some potential choices:

- ``sklearn.neighbors.KNeighborsRegressor`` (fast-ish, but often inaccurate
- ``sklearn.svm.SVR`` (potentially accurate, but slow)
- ``sklearn.ensemble.RandomForestRegressor`` (fast, and accurate with tuning)
- ``sklearn.ensemble.GradientBoostingRegressor`` (accurate with tuning, can be very slow depending on parameters)

Again, sub-sampling the data can help as you explore some of the slower techniques."

````python
from astroML.datasets import fetch_sdss_specgals
data = fetch_sdss_specgals()
````

````python
# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
````

First show the original one: linear regression

````python
# 1  Linear Regression
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
````

````python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print "As in the lecture, I see that it's not a good fit, as the rms is:", rms
rms_LR=rms
````

Try the next one on the list - K Neighbors regressor:

````python
# 2 K-Neighbors regression
from sklearn.neighbors import KNeighborsRegressor as KNR
est = KNR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print "For the KNR the rms is", rms
rms_KNR=rms
````

````python
print (rms_LR-rms_KNR)/(rms_LR) * 100.0
````

So I got an improvement by 28% over the Linear Regression. Try another one:

````python
# 3 Random Forest Regression
from sklearn.ensemble import RandomForestRegressor as RFR
est = RFR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
rms = np.sqrt(np.mean((y_test - y_pred)**2))
print rms
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.show()
print "For the RFR the rms is", rms
rms_RFR=rms
````

Wow! That looks very close to the best match. For certainty, try another one: 

````python
# 4 Gradient Boosting Regression
from sklearn.ensemble import GradientBoostingRegressor as GBR
est = GBR()
est.fit(X_train[::10], y_train[::10])
y_pred = est.predict(X_test)
rms = np.sqrt(np.mean((y_test - y_pred)**2))
print rms
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.show()
print "For the GBR the rms is", rms
rms_GBR=rms
````

Ouch, I see it gets dragged away on the top, so it would surely not be the best regressor, Compare all those with respect to the original Linear Regression:

````python
print "Linear regression", rms_LR, "reference, no improvement"
print "K-Neighbors regression", rms_KNR, "improvement of ", (rms_LR-rms_KNR)/(rms_LR) * 100.0 ,"%"
print "Random Forest Regression", rms_RFR, "improvement of ", (rms_LR-rms_RFR)/(rms_LR) * 100.0, "%"
print "Gradient Boosting Regression", rms_GBR, "improvement of ", (rms_LR-rms_GBR)/(rms_LR) * 100.0, "%"

````

So we can see that of all regressors, both K-Neighbors and Random Forest worked best.
