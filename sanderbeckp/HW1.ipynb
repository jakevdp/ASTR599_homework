```python
import numpy
%matplotlib inline
```

After importing numpy, I load the asteroid data in using numpy.genfromtxt and print the data to make sure it loaded correctly:

```python
data = numpy.genfromtxt('599/asteroids5000.csv', delimiter = ',')
print data
```

I then create a mask for the NaN values and check the mask

```python
mask = numpy.isnan(data)
mask
```

To see how many NaN values there are in the data, I apply the make to the data and take the length of the array:

```python
nmissingdata =len(data[mask])
print nmissingdata
```

To see how many complete rows there are, I apply the inverse mask and exclude any rows with False values. I then can check the length and print the data.

```python
completedata = data[~numpy.isnan(data).any(axis=1)]
print len(completedata)
print completedata
```

I can now find the maximum, minimum, mean, and standard deviation of the different data types by applying the appropriate numpy function along the columns.

```python
maximum = numpy.max(completedata, axis = 0)
minimum = numpy.min(completedata, axis = 0)
mean = numpy.mean(completedata, axis = 0)
stddev = numpy.std(completedata, axis = 0)
print maximum
print minimum
print mean
print stddev
```

To plot the semi-major-axis against the sin of the inclination, I convert the appropriate column from degrees to radians and apply the numpy.sin function and then plot with the semi-major-axis column

```python
sini = numpy.sin(completedata[:,2]*(numpy.pi/180.))
plt.scatter(sini,completedata[:,0])
```
