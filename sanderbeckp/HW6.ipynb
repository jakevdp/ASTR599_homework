First, import necessary packages and have matplotlib plots appear inline.

```python
import numpy
import matplotlib.pyplot as plt
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split
%matplotlib inline
```

Now we want to get our data and split it into random train and test subsets

```python
X, y = fetch_rrlyrae_combined()
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

We now plot the main sequence and RR Lyrae stars:

```python
N_plot = 5000
plt.scatter(X[-N_plot:,0],X[-N_plot:,1],c=y[-N_plot:],edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

Next, import the KNeighborsClassifier (takes the nearest neighbors and determines the calss of the unknown based on the class of the neighbors) and fit the model, then predict the class of each data point.

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print numpy.sum(y_pred == y_test)
print numpy.sum(y_pred!= y_test)
```

Calculate and print the precision, recall, f1-score of the results:

```python
from sklearn.metrics import classification_report
print classification_report(y_test,y_pred, target_names= ['MS star','RR Lyrae'])
```

Now do the same thing, but using RandomForestClassifier

```python
X, y = fetch_rrlyrae_combined()
X_train, X_test, y_train, y_test = train_test_split(X,y)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print numpy.sum(y_pred == y_test)
print numpy.sum(y_pred!= y_test)
```

We see that the precision and the f1-score improve, while the recall score declines.

```python
from sklearn.metrics import classification_report
print classification_report(y_test,y_pred, target_names= ['MS star','RR Lyrae'])
```

Now import our SDSS photometric redshifts:

```python
from astroML.datasets import fetch_sdss_specgals
from sklearn.cross_validation import train_test_split
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = fetch_sdss_specgals()
```

Get the appropriate data

```python
X = numpy.vstack([data['modelMag_%s' %f] for f in 'ugriz']).T
y = data['z']
```

Split data into random train and test subsets

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

Import LinearRegression, fit the model:

```python
from sklearn.linear_model import LinearRegression
```

```python
est = LinearRegression()
est.fit(X_train,y_train)
```

Now predict the linear fit

```python
y_pred = est.predict(X_test)
```

And plot the results

```python
plt.plot(y_test,y_pred,',k')
plt.plot([0,1],[0,1],':k')
plt.xlim(0,0.6)
plt.ylim(0,0.6)
```

We can see the fit is not very accurate

```python
rms = (numpy.mean((y_test - y_pred)**2))**(0.5)
print rms
```

Do this again, but using RandomForestRegressor:

```python
X = numpy.vstack([data['modelMag_%s' %f] for f in 'ugriz']).T
y = data['z']
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=10)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
plt.plot(y_test,y_pred,',k')
plt.plot([0,1],[0,1],':k')
plt.xlim(0,0.6)
plt.ylim(0,0.6)
```

The fit improves!

```python
rms = numpy.sqrt(numpy.mean((y_test - y_pred) ** 2))
print rms
```
