```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.lda import LDA
from sklearn.ensemble import RandomForestClassifier as RFC
import numpy as np
```

```python
#try SVC classifier 
clf = SVC()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

```python
#try RFC classifier 
clf = RFC()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

```python
#try GaussianNB classifier 
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

```python
#try LDA classifier 
clf = LDA()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)
from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

Scores for each classifier: RandomForestClassifier precision 0.82, recall 0.52, GassianNB precision 0.17, recall 0.83, LDA precision 0.16, recall 0.64. These all use the default settings, and should be compared to the K nearest neighbors classifier with n_neighbors=5, which yields precision 0.58 and recall 0.62. The biggest improvement seems to come from using RFC with n_estimators=10. It markedly increases the precision over the K nearest neighbors classifier, but decreases the recall by a bit. This indicates that it finds less false positives, but slightly increases the number of false negatives introduced.

## Part 2 of HW6

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
# now try the KNeighborsRegressor algorithm
from sklearn.neighbors import KNeighborsRegressor as KNR
est = KNR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k', color='blue')
plt.plot([0, 1], [0, 1], ':k', color='blue')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
# now try the SVR algorithm
from sklearn.svm import SVR
est = SVR()
est.fit(X_train[::5], y_train[::5])
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k', color='red')
plt.plot([0, 1], [0, 1], ':k', color='red')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
# now try the RFR algorithm
from sklearn.ensemble import RandomForestRegressor as RFR
est = RFR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k', color='green')
plt.plot([0, 1], [0, 1], ':k', color='green')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

```python
# now try the GBR algorithm
from sklearn.ensemble import GradientBoostingRegressor as GBR
est = GBR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k', color='grey')
plt.plot([0, 1], [0, 1], ':k', color='grey')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

The smallest RMS value comes from using the RandomForestRegressor, meaning the best fit comes from using this regressor (visually you can see that there is the least amount of scatter). The KNeighborsRegressor also provides a similarly good fit as can be seen from the relative lack of scatter. All the other regressors (using default values) do not seem to yield very good fits as compared to the two algorithms above--there are more serious deviations from the 1-to-1 line. I only ran the SVR algorithm on a subset of the data because it was taking a very long time.
