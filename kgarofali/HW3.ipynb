```python
%matplotlib inline
import astroML.datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get a variable star 
lcs = astroML.datasets.fetch_LINEAR_sample()
time, flux, dflux = lcs[1009459].T
```

```python
# plot a raw lightcurve 
plt.errorbar(time, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("MJD")
plt.ylabel("Mag")
plt.show()
```

```python
# use lomb scargle to find optimal period 
from astroML.time_series import lomb_scargle
periods = np.logspace(-2, 0, 10000, base=10)
periodogram = lomb_scargle(time, flux, dflux, omega=2 * np.pi / periods, generalized=True)
```

```python
# plot periodogram 
plt.plot(periods, periodogram)
plt.semilogx()
plt.xlabel("Period (days)")
plt.ylabel("Power")
plt.show()
```

```python
# can find peak of periodogram using argsort 
idx = np.argsort(periodogram)
print periods[idx[-1]], periodogram[idx[-1]]
```

```python
# fold the lightcurve to get the best period. find phase by MJD / period - integer portion MJD//period 
period = periods[idx[-1]]
phase = time / period - time // period 
print min(phase), max(phase)
```

```python
# now plot lightcurve using this phase. looks better! 
plt.errorbar(phase, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

```python
# define a method on a class to calculate our chi-square for a given model. class can have relevant
# variables. also want method to fit fourier model to our light curve 
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms):
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
```

```python
# create an instance of the class and plot it 
fourier = Fourier(phase, flux, dflux, 1)
mphase = np.arange(0, 1, 0.01)
model = fourier.evaluate(mphase, [10])
```

```python
plt.plot(mphase, model, "k-")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Model mag")
plt.show()
```

```python
# create a more complex instance of the class and plot 
fourier = Fourier(phase, flux, dflux, 2)
mphase = np.arange(0, 1, 0.01)
model = fourier.evaluate(mphase, [10, 0.1, 0.0])
```

```python
plt.plot(mphase, model, "k-")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Model mag")
plt.show()
```

```python
# use optimization routine fmin_bfgs from scipy as our minimization routine 
import scipy.optimize as opt
```

```python
# define method on class to initiliaze our guesses for the model. inherit from main Fourier class. 
class Fourier2(Fourier):
    def __init__(self, phase, flux, dflux, nterms):
        Fourier.__init__(self, phase, flux, dflux, nterms)
                
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess
```

```python
# now we can plot for a range of models (diff num of terms) to see how chi square improves 
nterms = np.arange(1, 10)
chi2   = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
plt.plot(nterms, chi2)
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.show()
```

```python
# can also look at delta chi square. it goes to zero which means less and less improvement for each
# subsequent addition of new terms to the model 
delta_chi2 = chi2[:-1] - chi2[1:]
plt.plot(nterms[1:], delta_chi2)
plt.xlabel("N terms")
plt.ylabel("Delta chi2")
plt.show()
```

```python
# we can also look at penalties imposed by BIC/AIC. plot them to see how they compare to chi square 
# BIC = chi2 + Nterm * ln(Ndata)
# AIC = chi2 + Nterm * 2 
nterms     = np.arange(1, 30, 2)
chi2       = []
BICpenalty = []
AICpenalty = []
best_fit = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    BICpenalty.append(len(guess) * np.log(len(fourier.phase)))
    AICpenalty.append(len(guess) * 2.0)
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
    best_fit.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[0])
best_fit = np.array(best_fit)
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
delta_chi2 = chi2[:-1] - chi2[1:]
red_chi2 = (chi2 / (len(flux) - len(best_fit) - 1))

# now plot AIC and BIC as function of number of terms 
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + BICpenalty, label="BIC")
plt.plot(nterms, chi2 + AICpenalty, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()

# plot delta_chi2 as a function of number of terms 
plt.plot(nterms[1:], delta_chi2)
plt.xlabel("N terms")
plt.ylabel("Delta chi2")
plt.show()
```

```python
# see which fits have minimum BIC, AIC, and delta_chi2 and find which order model that suggests 
print "Minimum BIC: %i, Minimum AIC: %i, Minimum delta_chi2: %i" % (np.argsort(chi2+BICpenalty)[0],np.argsort(chi2+AICpenalty)[0], np.argsort(delta_chi2)[0])
print "Suggesting, respectively, model orders: %i, %i, %i" % (nterms[14],nterms[14],nterms[8]) 
# note for above, min delta_chi2 corresponds to most improvement from one model to the next, so
# we want nterms[min(delta_chi2 +1)] to get the best model. 
```

From this anaylsis we would pick model order 29 from BIC/AIC values, and model order 17 from the delta_chi2 values. BIC/AIC tells us when to stop in the following way: we want the improvement in chi2 to be smaller than the penalty imposed by adding more terms (overall penalty is chi2 +BIC/AIC). The delta_chi2 tells us how much our chi2 value improves from one model to the next. We want to find the minimum of this, because that tells us when we have "maximized" our improvement from one model to the next given the addition of new terms to the model. The delta_chi2 value approaches zero, meaning that the improvement in chi2 from one model to the next becomes more and more insignificant with the addition of more terms. By finding the minimum of delta_chi2 we are finding the model that corresponds to a point before which improvement comes at the expense of addition of new terms. 

```python
#plot up best fit models given AIC/BIC and delta_chi2\n
fourier_AIC_BIC = Fourier2(phase,flux,dflux, nterms[14])
guess_AIC_BIC = fourier_AIC_BIC.makeGuess()
result_AIC_BIC = opt.fmin_bfgs(fourier_AIC_BIC.chi2, x0=[guess_AIC_BIC], disp=0, full_output=1)
bestfit_AIC_BIC = result_AIC_BIC[0]

fourier_delta = Fourier2(phase,flux,dflux, nterms[8])
guess_delta = fourier_delta.makeGuess()
result_delta = opt.fmin_bfgs(fourier_delta.chi2, x0=[guess_delta], disp=0, full_output=1)
bestfit_delta = result_delta[0]

plt.errorbar(phase, flux, yerr=dflux, fmt='ro')
plt.plot(mphase, fourier_AIC_BIC.evaluate(mphase, bestfit_AIC_BIC), 'g-', label="Best fit AIC/BIC")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.legend()
plt.show()

plt.errorbar(phase, flux, yerr=dflux, fmt="ro")
plt.plot(mphase, fourier_AIC_BIC.evaluate(mphase, bestfit_AIC_BIC), "b-", label="Best fit Delta_chi2")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.legend()
plt.show()
```
