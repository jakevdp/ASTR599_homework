# Homework 4: Generate image of a disk galaxy and save as FITS file.

First, import relevant modules.

```python
import astropy.io.fits as fits
import numpy as np
import matplotlib.pyplot as plt
import pylab
%pylab inline
```

Generate grid of points and multiple exponentially decaying disk components. Plot up galaxy image.

```python
x, y = np.ogrid[-50:50, -50:50]
r = np.sqrt(x**2 + y**2)

n1 = 5*np.exp(-r/10.)
n2 = 2*np.exp(-r/50.) 
n3 = np.exp(-r/150.)
n = n1+n2+n3

imshow(n)
```

Save the galaxy image as a .fits file - FITS image generated!

```python
hdu = fits.PrimaryHDU(n)
hdu.writeto('DiskGalaxy.fits')
```
