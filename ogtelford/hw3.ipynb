Generate the periodogram of the data.

```python
from astroML.time_series import lomb_scargle
periods = np.logspace(-1, 0, 100000, base=10) 
periodogram = lomb_scargle(time, flux, dflux, omega=2 * np.pi / periods, generalized=True)

plt.plot(periods, periodogram)
plt.semilogx()
plt.xlabel("Period (days)")
plt.ylabel("Power")
plt.show()
```

Plot up the lightcurve for this variable star.

```python
idx = np.argsort(periodogram)
print periods[idx[-1]], periodogram[idx[-1]]

period = periods[idx[-1]]
phase = time / period - time // period # // returns the integer portion of a fraction - only want the fraction part here
print min(phase), max(phase) 

plt.errorbar(phase, flux, yerr=dflux, fmt="ro") #changing to finer sampling gave much better lightcurve!
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

Define a class to model the data, calculate the chi-squared, and generate guesses for optimization.

```python
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms):
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
    
class Fourier2(Fourier):
    def __init__(self, phase, flux, dflux, nterms):
        Fourier.__init__(self, phase, flux, dflux, nterms)
                
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess

import scipy.optimize as opt
```

Fit models with varying number of terms to the data; for each, calculate chi-squared, AIC and BIC penalties.

```python
nterms     = np.arange(1, 30, 2)
chi2       = []
BICpenalty = []
AICpenalty = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    BICpenalty.append(len(guess) * np.log(len(fourier.phase)))
    AICpenalty.append(len(guess) * 2.0)
    chi2.append(opt.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
```

```python
print chi2
print BICpenalty
print AICpenalty
```

Generate plot of the delta chi-squared as a function of number of terms in the model. We see that the decrease in chi-squared is most significant for increasing the number of terms when nterms is small; the delta chi-squared gets much smaller as we get to higher-order models.

```python
delta_chi2 = chi2[:-1] - chi2[1:]
plt.plot(nterms[1:], delta_chi2)
plt.xlabel("N terms")
plt.ylabel("Delta chi2")
plt.show()
```

Generate plot of chi-squared, chi-squared + AIC, and chi-squared + BIC as a function of number of terms in the model.

```python
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + BICpenalty, label="BIC")
plt.plot(nterms, chi2 + AICpenalty, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()
```

Use the minimum chi-squared, chi-squared + AIC, and chi-squared + BIC values to select the best fit for the model without overfitting (using too many terms).

```python
print np.argsort(chi2)[0],np.argsort(chi2+BICpenalty)[0],np.argsort(chi2+AICpenalty)[0] 
print nterms[14], nterms[5], nterms[6] #find the model order at which chisq, chisq+penalties are minimized
```

### Solution: The minimum chi-squared indicates that the highest order model (29 terms) is the best. Adding the two penalties gives different results; the BIC penalty gives 11 terms as the optimum, while the AIC penalty gives 13 terms. 

Verify that fits selected using above metrics (here the BIC penalty) actually give reasonable fits to the data.

```python
fourier11 = Fourier2(phase, flux, dflux, 11)
guess = fourier11.makeGuess()
result11 = opt.fmin_bfgs(fourier11.chi2, x0=[guess], disp=0, full_output=1) 
best_fit11 = result11[0]
mphase = np.arange(0, 1, 0.01)

plt.errorbar(fourier11.phase, fourier11.flux, yerr=fourier11.dflux, fmt="ro")
#plt.plot(mphase, fourier11.evaluate(mphase, guess), "b--", label="Initial guess")
plt.plot(mphase, fourier11.evaluate(mphase, best_fit11), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.title("Comparing Best Fit Model to the Lightcurve")
plt.show()
```
