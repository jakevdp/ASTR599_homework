## Exercise 1: RR Lyrae Classification

```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
import matplotlib.pyplot as plt
%pylab inline

N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
import numpy as np
#from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
#from sklearn.lda import LDA
#clf = GaussianNB()
clf = RandomForestClassifier(n_estimators = 20)
#clf = LDA()

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print np.sum(y_pred == y_test)
print np.sum(y_pred != y_test)

from sklearn.metrics import classification_report
print classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae'])
```

When using sklearn.svm.SVC, get following message:

sklearn.naive_bayes.GaussianNB: precision = 0.16, recall = 0.86

sklearn.lda.LDA: precision = 0.14, recall = 0.63

sklearn.ensemble.RandomForestClassifier: precision = 0.84, recall = 0.53 (varies a lot between runs); tried varying n_estimators, criterion, and max_features - got no better results than default.

#### Result: Best combination of precision/recall for RR Lyrae data given by RandomForestClassifier with default values: precision = 0.84, recall = 0.53. Better recall of 0.86 is obtained with GaussianNB, but the precision is only 0.16.

## Exercise 2: Photometric Redshift

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
print np.shape(X)
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
#from sklearn.svm import SVR
#est = KNeighborsRegressor(n_neighbors = 15)
est = RandomForestRegressor(n_estimators = 20, max_features=None)
#est = SVR()
est.fit(X_train, y_train) 
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

sklearn.neighbors.KNeighborsRegressor: rms deviation = 0.0233; visually much better fit (high-z end fits 1-1 relation well). Increase n_neighbors to 15 gives rms = 0.0226. Changing leaf_size has no effect.

sklearn.ensemble.RandomForestRegressor: rms = 0.0229; changing n_estimators to 20 gives 0.0225; changing max_features had no effect.

sklearn.svm.SVR: rms = 0.0418; visually not a great fit.

#### Result: best fit is obtained using RandomForestRegression(n_estimators=20); the rms deviation = 0.0225.
